Additional regex syntax - OR | operator

```python
re.search(r'man|woman','A man was here in China')
```

```python
re.search(r'man|woman','A woman was also here in SH')
```

```python
re.search(r'man|woman','A man and a woman were both here in Jingan.')
```

wildcard character 通配符 .

```python
re.findall(r'.at','A fat cat sat on the mat')
```

```python
re.findall(r'.at','The bat went splat.')
```

one or more non-whitespace words that end with 'at'
\S= non-whitespace, += more letters

```python
re.findall(r'\S+at','The bat went splat')
```

find start(^) or end($) patterns

```python
re.findall(r'\d$','This text ends with a number 3')
```

```python
re.findall(r'^\d','1 is the lonliest number.')
```

how exclusions work

```python
re.findall(r'[^\d]','There are 3 numbers here 123 and 345 in the sentence')
```

```python
re.findall(r'[^\d]+','There are 3 numbers here 123 and 345 in the sentence')
```

```python
text='This is a text with lots of punctuations, comma and period. Really!! '
re.findall(r'[^!.?.,]+',text)
```

```python
clean_text=''.join(re.findall(r'[^!.?.,]+',text))
clean_text
```
